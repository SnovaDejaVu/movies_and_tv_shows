# Movies and TV shows

### 1. Import pandas module

```python
import pandas as pd
from datetime import datetime
```

### 2. Importing dataset

```python
tv_shows = pd.read_csv("Movies_titles.csv")
```

### 5. Examining missing values

```python
print("Missing values: ")
tv_shows.isnull().mean()
```

### 6. Check datatype in each column

```python
print("Column datatypes: ")
print(tv_shows.dtypes)
```

### 7. Getting all the columns with string/mixed type values and removing leading and trailing characters from columns with str type

```python
str_cols = list(tv_shows.columns)
str_cols.remove('release_year')

for i in str_cols:
    tv_shows[i] = tv_shows[i].str.strip()
```

### 8. Replace NaN values

```python
columns = ['director', 'cast', 'country', 'rating', 'date_added']

for column in columns:
    tv_shows[column] = tv_shows[column].fillna("")
```

### 9. Separate the information into 2 other columns and filter out the rows with NaN values and replace them with 0

```python
rows = []
for i in range(len(tv_shows)):
    if tv_shows['date_added'].iloc[i] == "":
        rows.append(i)

        
month_added = []
year_added = []
for i in range(len(tv_shows)):
    if i in rows:
        month_added.append(0)
        year_added.append(0)
    else:
        date = tv_shows['date_added'].iloc[i].split(" ")
        month_added.append(date[0])
        year_added.append(int(date[2]))
        

for i, month in enumerate(month_added):
    if month != 0:
        datetime_obj = datetime.strptime(month, "%B")
        month_number = datetime_obj.month
        month_added[i] = month_number

        
tv_shows.insert(7, "month_added", month_added, allow_duplicates = True)
tv_shows.insert(8, "year_added", year_added, allow_duplicates = True)

tv_shows = tv_shows.drop(columns='date_added')

tv_shows.head(3)
```
